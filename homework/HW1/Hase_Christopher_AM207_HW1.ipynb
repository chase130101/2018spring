Let $d$ be the event that a randomly selected child has Type I diabetes.

Let $p$ be the event that a randomly selected child tests positive for Type I diabetes.

$P(d) = 0.00228$

$P(p|d^c) = 0.01$

$P(p^c|d) = 0$

$P(d|p)=\dfrac{P(p|d) \cdot P(d)}{P(p|d) \cdot P(d) + P(p|d^c) \cdot P(d^c)}

=\dfrac{1 \cdot 0.00228}{1 \cdot 0.00228 + 0.01 \cdot 0.99772}

Part A

First we need to check that $f_X(x)$ is a valid PDF.

Since $\lambda_0>0$, $\lambda_1>0$, and $f_X$ is only supported on the nonnegative quadrant of $\mathbb{R}^2$, $f_X\geq0$ over its support.

$1 = \displaystyle\int_0^\infty\displaystyle\int_0^\infty 4\lambda_1^2x_1x_2\exp\big(-\lambda_0(x_1^2+x_2^2)\big)dx_1dx_2=4\lambda_1^2\displaystyle\int_0^\infty x_1\exp(-\lambda_0x_1^2)dx_1\displaystyle\int_0^\infty x_2\exp(-\lambda_0x_2^2)dx_2=4\lambda_1^2\cdot\dfrac{0-1}{-2\lambda_0}\cdot\dfrac{0-1}{-2\lambda_0}=\dfrac{\lambda_1^2}{\lambda_0^2}\Rightarrow \lambda_0^2=\lambda_1^2$

Since $\lambda_0>0$ and $\lambda_1>0$, $\lambda_0=\lambda_1$. That is, for $f_X$ to be a valid PDF, we must have $\lambda_0=\lambda_1$.

Then, we can write that $f_X(\mathbf{x})=4\lambda_0^2x_1x_2\exp\big(-\lambda_0(x_1^2+x_2^2)\big)$.

$L(\lambda_0|x^{(0)},x^{(1)},x^{(2)},x^{(3)})=\displaystyle\prod_{i=0}^3f_X(\mathbf{x}^{(i)})=4^4\lambda_0^8(0.5)(2.5)(3.2)(1.3)(2.72)(5.84)(10.047)(0.354)\exp\big(-\lambda_0(0.5^2+2.5^2+3.2^2+1.3^2+2.72^2+5.84^2+10.047^2+0.354^2)\big)\\
=75208.1186\cdot\lambda_0^8\exp(-161.001525\lambda_0)$

$l(\lambda_0|x^{(0)},x^{(1)},x^{(2)},x^{(3)})=\log\big(L(\lambda_0|x^{(0)},x^{(1)},x^{(2)},x^{(3)})\big)=\log(75208.1186)+8\log(\lambda_0)-161.001525\cdot\lambda_0$

$\dfrac{dl}{d\lambda_0}=\dfrac{8}{\lambda_0}-161.001525$

$\dfrac{8}{\hat\lambda_0}-161.001525\equiv 0\Rightarrow \hat\lambda_0 = \dfrac{8}{161.001525}$

$\dfrac{d^2l}{d\lambda_0^2}=\dfrac{-8}{\lambda_0^2}<0$ $\forall\lambda_0$, so $L(\lambda_0|x^{(0)},x^{(1)},x^{(2)},x^{(3)})$ is maximized at $\hat\lambda_0$.

For completion, we can write that $\hat\lambda_1 = \dfrac{8}{161.001525}$ also.

```python
def get_bootstrap_sample(dataset):
    return np.random.choice(dataset, size = len(dataset), replace = True) # sample with replacement

def perform_bootstrap(dataset, num_samples):
    samples_matrix = np.zeros((num_samples, len(dataset))) # store samples in a matrix
    for i in range(num_samples):
        samples_matrix[i, :] = get_bootstrap_sample(dataset) # use get_bootstrap_sample function
        
    return samples_matrix
```
